```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
temps = "https://raw.githubusercontent.com/roualdes/data/master/temperature.csv"
```

```python
df = pd.read_csv(temps)
df
```

```python
plt.hist(df["hilo"], bins = "fd", histtype = "step" )
```

*Bin doesn't work as the way it decides the number of bars and their width loses large amounts of data between bars*

```python
plt.hist(df["hilo"], histtype = "step" )
plt.title("Histogram of Hilo")
```

*This shows a far more accurate representation of the 'large' dataset*

One interesting thing about the histogram of the Hilo temperatures is how matplot chose the y-values as a count of records
at the designated temperature. This depicts a normal distribution of data.

```python
plt.scatter(df["hilo"], df["death_valley"])
plt.xlabel("Hilo")
plt.ylabel("Death_Valley_Temperatures")
plt.title("Scatter Plot of Temp (Death_Valley/Hilo)")
```

Using this graph we can see that both locations follow a similar trend and have more counts towards the middle of the data showing a normal distribution. This graph does not provide enough information for a causal relationship claim. This is because without a time scale this data simply shows that both sets have a mean value, for which most of the counts surround. A variable of time is necessary to depict whether or not the data is a causal relationhsip, or simply two sets of data with a similar distribution.
